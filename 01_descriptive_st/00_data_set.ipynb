````python
df = sns.load_dataset('tips')
````

````python
df
````

### column を追加する　→　tip_rate
- tip(チップ)　$\div$　total_bill(合計請求金額)

    - お会計に対して　→ 何%チップを支払っているか？

````python
df['tip_rate'] = df['tip'] / df['total_bill']
````
